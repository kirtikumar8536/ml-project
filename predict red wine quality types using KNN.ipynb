```python
import pandas as pd

path = "C:/Users/kirti/Desktop/intern_project/2022project_1/KNN/Winequalitydatatset - 2/winequalityred.csv"

data = pd.read_csv(path) #reading the csv file

#print(data)
#print(data.info())
print(data.isnull().sum())
print("*"*90)
mean1 = data['fixed acidity'].median()
print(mean1)
data['fixed acidity'] = data['fixed acidity'].fillna(mean1)


print("*"*90)

mean2 = data['volatile acidity'].median()
print(mean2)
data['volatile acidity'] = data['volatile acidity'].fillna(mean2)


print("*"*90)

mean3 = data['citric acid'].median()
print(mean3)
data['citric acid'] = data['citric acid'].fillna(mean3)


print("*"*90)
mean4 = data['pH'].median()
print(mean4)
data['pH'] = data['pH'].fillna(mean4)


print("*"*90)
mean5 = data['sulphates'].median()
print(mean5)
data['sulphates'] = data['sulphates'].fillna(mean5)
print("*"*90)

inputs = data.drop(['type','quality'],'columns')
output = data.drop(['type','fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol'],'columns')

import sklearn
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(inputs,output,test_size=0.2)

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5) #intializing the model

model.fit(x_train,y_train) #train the model

y_pred = model.predict(x_test)
#print(y_pred)
#print(y_test)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)
print(cm)

res = model.predict([[7.4,0.7,0,1.9,0.076,11,34,0.9978,3.5,0.56,9.4]])
print(res)

```
